```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'salary.csv')
df.head()
```

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = df.iloc[:,:-1]
calc_vif(X)
```

```python
# VIF are showing more because scaling of data is not done here .
```

```python
X = df.drop(['Age','salary'],axis=1)
calc_vif(X)
```

```python
df2 = df.copy()
df2['Age_at_joining'] = df.apply(lambda x: x['Age'] - x['Year of service'],axis=1)
X = df2.drop(['Age','Year of service','salary'],axis=1)
calc_vif(X)
```
